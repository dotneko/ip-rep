```python
import urllib.request
import os.path
import pandas as pd

url = "http://reputation.alienvault.com/reputation.data"

try:
    response, headers = urllib.request.urlretrieve(url)
    av = pd.read_csv(response, sep="#")
    av.columns = ["IP", "Reliability", "Risk", "Type", "Country", "Locale", "Coords", "x"]
except:
    print("Error occurred")
    
```

```python

country_ct = pd.value_counts(av['Country'])
print(country_ct[:25].index)
print(country_ct[:25].value_counts)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

data = [go.Bar(
            x=country_ct[:20].index,
            y=country_ct[:20],
            name='Summary by Country'
    )]

py.iplot(data, filename='avip-country-bar')
```
